```python
def count_repeated_tokens(seq1: str, seq2: str):
    # Tokenize by splitting on whitespace
    tokens1 = seq1.strip().split()
    tokens2 = seq2.strip().split()

    # Convert to sets to find repeated tokens
    set1 = set(tokens1)
    set2 = set(tokens2)

    # Find common tokens
    common_tokens = set1.intersection(set2)

    # Count how many times each common token appears in both sequences
    repeated_count = 0
    for token in common_tokens:
        count_in_seq1 = tokens1.count(token)
        count_in_seq2 = tokens2.count(token)
        repeated_count += min(count_in_seq1, count_in_seq2)

    return {
        "repeated_count": repeated_count,
        "total_tokens_seq1": len(tokens1),
        "total_tokens_seq2": len(tokens2),
    }
```

# calculate the scores and/or load the scores

```python
def calculate_fitness(row):
    if not isinstance(row, list):
        row = [row]
    assert len(row) == 1
    batch_scores = score_sequences(
        row,
        model,
        tokenizer,
        device=device,
    )
#     print(batch_scores)
    print(batch_scores)
    return [i.item() for i in batch_scores][0]
```

```python
data_list = {}
data_list_corr = {}
print(model_name)
for species in ["ecoli_protein_dna_dms-1", "other_prokaryotic_dna_dms-1"]:
    for study_name in tasks[species]:
        print(study_name)
        data_dna = pd.read_csv(f"{data_dir}/{species}/{study_name}_dms_dna.csv")
        if os.path.exists(f"{data_dir}/{species}/{study_name}_dms_dna_with_predicted.csv"):
            data_dna = pd.read_csv(f"{data_dir}/{species}/{study_name}_dms_dna_with_predicted.csv")
        if not model_name in data_dna.columns:
            data_dna[model_name] = data_dna['mutated_sequence_dna'].apply(calculate_fitness)
            data_dna.to_csv(f"{data_dir}/{species}/{study_name}_dms_dna_with_predicted.csv", index=False)
            print(data_dna)

        data_dna = data_dna.dropna()
        data_list[study_name] = data_dna
        data_list_corr[study_name] = {}
        for column in tasks_ground_truth_name[species][study_name]:
            corr = spearmanr(data_dna[column], data_dna[model_name]).correlation
            print(column, corr)
            data_list_corr[study_name][column] = corr
```

## Sequence Generation

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import torch
from transformers import (
  AutoTokenizer,
  EsmForProteinFolding,
  set_seed,
  T5Tokenizer,
  AutoConfig,
  AutoModelForCausalLM
)
from Bio import SeqIO, AlignIO, Phylo
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Align.Applications import MuscleCommandline
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
import matplotlib.pyplot as plt
```

```python
#Load Evo Model
model_name = 'togethercomputer/evo-1-131k-base'
config = AutoConfig.from_pretrained(model_name, trust_remote_code=True, revision="1.1_fix")
config.use_cache = True
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    config=config,
    trust_remote_code=True,
    revision="1.1_fix"
)
tokenizer = AutoTokenizer.from_pretrained(model_name,trust_remote_code = True)
```

 Our prompting scheme follows "Greengenes-style lineage strings", which concatenate all taxa starting with the most ancestral and ending with the most current, separated by semicolons. A single character prefix is also added to each taxon indicating its rank.
For more information, see: https://github.com/qiyunzhu/woltka/blob/main/doc/hierarchy.md#3-lineage-strings---lineage

```python
prompt = ["|d__Bacteria;p__Tenericutes;c__Mollicutes;o__Mycoplasmatales;f__Mycoplasmataceae;g__Mycoplasma;s__Mycoplasma genitalium||"]
#Following are a few more species prompts for generation:
#"|d__Bacteria;p__Bacillota;c__Bacilli;o__Staphylococcales;f__Staphylococcaceae;g__Staphylococcus;s__Staphylococcus aureus||",
#"|d__Bacteria;p__Pseudomonadota;c__Gammaproteobacteria;o__Enterobacterales;f__Enterobacteriaceae;g__Klebsiella;s__Klebsiella pneumoniae||",
#"|d__Bacteria;p__Pseudomonadota;c__Gammaproteobacteria;o__Enterobacterales;f__Enterobacteriaceae;g__Escherichia;s__Escherichia||"
device = device ="cuda" if torch.cuda.is_available() else "cpu"
input = torch.tensor(tokenizer(prompt)['input_ids']).to(device)
```

```python
model.to(device)
input=input.to(model.device)
```

```python
# ↓ A few more variables to control the generation process
max_tokens = 1024 # we want to generate maximally 1024 new tokens
temperature = 1. # The sampling temperature (between 0 and 2) controls
# the randomness with which the model picks tokens during generation.
# Higher values (eg, >1) will make the generations more diverse (-> more random).
# Lower values (eg, <1) will make them less diverse (-> more deterministic).
top_k = 4 # Specifies the maximum number of tokens to be considered at each
# generation step, based on their probability of occurrence.
# This can help speed up the generation process and improve
# the quality of the generated tokens by focusing on the most likely options.
top_p = 1. # As an alternative to the sampling temperature, we can also set
# the "top_p" parameter. The model only considers the those tokens for
# generation that contain the top_p probability mass.
# Eg, with top_p = 0.1, the model only considers the tokens comprising
# the top 10% probability -> leading to generally less diverse generations.
# When top_p = 1, all tokens are considered -> more diverse generations.
```

```python

output = model.generate(input,max_length = max_tokens,
  temperature = temperature,
  top_k = top_k,
  top_p = top_p,
  do_sample=True
  )
```

```python
output_seqs = tokenizer.decode(output)
```

```python
sequence_fasta='test.fasta'
gen_dna_seq = output_seqs.split('||')[1]
print(gen_dna_seq)
gen_dna_seq_record = SeqRecord(
    Seq(gen_dna_seq),
    id="evo-dna",
    description="DNA sequence generated by Evo."
    )
with open(sequence_fasta, "w") as output_handle:
    SeqIO.write(gen_dna_seq_record, output_handle, "fasta")
```

```python
fasta_files = ["/content/k_pneumoniae.fasta", "/content/s_aureus.fasta", "/content/e_coli.fasta","/content/m_genitalium.fasta"]
generated_sequences = {}

# Loop through all .fasta files
for filename in fasta_files:
    species_name = os.path.splitext(filename)[0].split('/')[-1]
    print(species_name)
    record = next(SeqIO.parse(filename, "fasta"))  # assumes one record per file
    generated_sequences[species_name] = str(record.seq)
```

```python
from itertools import product
from collections import Counter
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

#Tetranucleotide usage deviations
def generate_kmers(k=4):
    return [''.join(p) for p in product('ACGT', repeat=k)]

# === STEP 2: Compute normalized 4-mer frequency vectors ===
def compute_kmer_freq(seq, k=4):
    counts = Counter([seq[i:i+k] for i in range(len(seq) - k + 1)])
    total = sum(counts.values())
    return {kmer: counts.get(kmer, 0) / total for kmer in generate_kmers(k)}

# === STEP 3: Create matrix from all sequences ===
def build_feature_matrix(sequences, k=4):
    labels = list(sequences.keys())
    matrix = [list(compute_kmer_freq(sequences[label], k).values()) for label in labels]
    return np.array(matrix), labels

# === STEP 4: Compute distances and generate linkage matrix ===
def compute_linkage(matrix, method='average', metric='euclidean'):
    return linkage(matrix, method=method, metric=metric)

# === STEP 5: Draw dendrogram ===
def draw_dendrogram(linkage_matrix, labels, title="TUD-based Phylogenetic Tree"):
    plt.figure(figsize=(10, 5))
    # Set orientation to 'left' for a horizontal dendrogram
    dendrogram(linkage_matrix, labels=labels, orientation='left')
    plt.title(title)
    plt.tight_layout()
    plt.show()

```

```python
# === RUN ===
matrix, labels = build_feature_matrix(generated_sequences, k=4)

# Apply log2 transformation with a small pseudocount to avoid log(0)
log_matrix = np.log2(matrix + 1e-6)

# Compute linkage using Euclidean distances on the log2-transformed matrix
linkage_matrix = compute_linkage(log_matrix, method='average', metric='euclidean')

draw_dendrogram(linkage_matrix, labels)
```
