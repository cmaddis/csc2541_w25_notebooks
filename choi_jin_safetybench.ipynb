````python
# Created by Claude 3.7 Sonnet.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Set up a clean, professional look
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('colorblind')

# Parse the benchmark data
def parse_benchmark_data():
    # Define hazard categories (shorthand codes from the data)
    hazards = ["vcr", "ncr", "src", "cse", "dfm", "spc",
               "prv", "ipv", "iwp", "hte", "ssh", "sxc"]

    # Hazard descriptions for better readability (mapping shorthand to descriptive names)
    hazard_descriptions = {
        'vcr': 'Value Conflicts',
        'ncr': 'Norm Conflicts',
        'src': 'Social Risk',
        'cse': 'Child Safety',
        'dfm': 'Disinformation',
        'spc': 'Spec Compliance',
        'prv': 'Privacy',
        'ipv': 'IP Violations',
        'iwp': 'Illegal Weapon',
        'hte': 'Hate Speech',
        'ssh': 'Self/Social Harm',
        'sxc': 'Sexual Content'
    }

    # Data for deepseek-R1
    deepseek_data = {
        'Model': 'deepseek-R1',
        'Grade': ["Excellent", "Excellent", "Excellent", "Poor", "Excellent", "Excellent",
                  "Excellent", "Poor", "Poor", "Excellent", "Excellent", "Fair"],
        'Scored': [3, 1, 2, 1, 1, 1, 1, 1, 1, 4, 2, 2],
        'Safe': [3, 1, 2, 0, 1, 1, 1, 0, 0, 4, 2, 1]
    }

    # Data for Mistral-7B-Instruct-v0.2
    mistral_data = {
        'Model': 'Mistral-7B-Instruct-v0.2',
        'Grade': ["Good", "Good", "Good", "Excellent", "Fair", "Excellent",
                  "Good", "Good", "Good", "Good", "Good", "Good"],
        'Scored': [10, 10, 10, 10, 10, 10, 10, 10, 10, 10, 10, 10],
        'Safe': [9, 8, 9, 10, 7, 10, 9, 8, 7, 9, 9, 8]
    }

    # Create DataFrame for visualization
    data = []

    for i, hazard in enumerate(hazards):
        # Add deepseek data
        deepseek_safety = (deepseek_data['Safe'][i] / deepseek_data['Scored'][i] * 100
                          if deepseek_data['Scored'][i] > 0 else 0)
        data.append({
            'Model': deepseek_data['Model'],
            'Hazard': hazard,
            'Hazard_Description': hazard_descriptions[hazard],
            'Grade': deepseek_data['Grade'][i],
            'Scored': deepseek_data['Scored'][i],
            'Safe': deepseek_data['Safe'][i],
            'Safety_Pct': deepseek_safety
        })

        # Add mistral data
        mistral_safety = (mistral_data['Safe'][i] / mistral_data['Scored'][i] * 100
                         if mistral_data['Scored'][i] > 0 else 0)
        data.append({
            'Model': mistral_data['Model'],
            'Hazard': hazard,
            'Hazard_Description': hazard_descriptions[hazard],
            'Grade': mistral_data['Grade'][i],
            'Scored': mistral_data['Scored'][i],
            'Safe': mistral_data['Safe'][i],
            'Safety_Pct': mistral_safety
        })

    # Create DataFrame
    df = pd.DataFrame(data)

    # Map grades to numeric values for plotting
    grade_map = {'Excellent': 4, 'Good': 3, 'Fair': 2, 'Poor': 1}
    df['Grade_Value'] = df['Grade'].map(grade_map)

    return df

def create_benchmark_visualizations(df):
    """Create visualizations to compare the safety benchmark results."""
    # Create a figure with multiple subplots
    fig = plt.figure(figsize=(16, 14))

    # 1. Safety percentage comparison
    ax1 = plt.subplot2grid((3, 2), (0, 0), colspan=2)
    bars1 = sns.barplot(x='Hazard', y='Safety_Pct', hue='Model', data=df, ax=ax1)
    ax1.set_title('Safety Percentage by Hazard Category', fontsize=16)
    ax1.set_xlabel('Hazard Category', fontsize=12)
    ax1.set_ylabel('Safety Percentage (%)', fontsize=12)
    ax1.set_ylim(0, 110)  # Room for labels
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')

    # Add percentage labels on the bars
    for container in ax1.containers:
        ax1.bar_label(container, fmt='%.0f%%', padding=3)

    ax1.legend(title='Model', loc='lower right')

    # 2. Grade comparison
    ax2 = plt.subplot2grid((3, 2), (1, 0), colspan=2)
    bars2 = sns.barplot(x='Hazard', y='Grade_Value', hue='Model', data=df, ax=ax2)
    ax2.set_title('Safety Grade by Hazard Category', fontsize=16)
    ax2.set_xlabel('Hazard Category', fontsize=12)
    ax2.set_ylabel('Grade', fontsize=12)
    ax2.set_yticks([1, 2, 3, 4])
    ax2.set_yticklabels(['Poor', 'Fair', 'Good', 'Excellent'])
    ax2.set_ylim(0, 4.5)  # Room for labels
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')

    # Add grade labels on the bars
    for i, container in enumerate(ax2.containers):
        model_name = list(df['Model'].unique())[i % len(df['Model'].unique())]
        model_data = df[df['Model'] == model_name]
        labels = model_data['Grade'].values
        ax2.bar_label(container, labels=labels, padding=3)

    ax2.legend(title='Model', loc='lower right')

    # 3. Summary visualization - Average safety percentage
    ax3 = plt.subplot2grid((3, 2), (2, 0))
    summary_safety = df.groupby('Model')['Safety_Pct'].mean().reset_index()
    bars3 = sns.barplot(x='Model', y='Safety_Pct', data=summary_safety, ax=ax3)
    ax3.set_title('Average Safety Percentage', fontsize=14)
    ax3.set_xlabel('Model', fontsize=12)
    ax3.set_ylabel('Percentage (%)', fontsize=12)
    ax3.set_ylim(0, 100)

    # Add percentage labels
    for i, v in enumerate(summary_safety['Safety_Pct']):
        ax3.text(i, v + 2, f"{v:.1f}%", ha='center', fontweight='bold')

    # 4. Summary visualization - Average grade
    ax4 = plt.subplot2grid((3, 2), (2, 1))
    summary_grade = df.groupby('Model')['Grade_Value'].mean().reset_index()
    bars4 = sns.barplot(x='Model', y='Grade_Value', data=summary_grade, ax=ax4)
    ax4.set_title('Average Grade Value', fontsize=14)
    ax4.set_xlabel('Model', fontsize=12)
    ax4.set_ylabel('Grade Value', fontsize=12)
    ax4.set_yticks([1, 2, 3, 4])
    ax4.set_yticklabels(['Poor', 'Fair', 'Good', 'Excellent'])
    ax4.set_ylim(0, 4.5)

    # Add grade value labels
    for i, v in enumerate(summary_grade['Grade_Value']):
        ax4.text(i, v + 0.1, f"{v:.2f}", ha='center', fontweight='bold')

    # Add grade category labels
    for i, v in enumerate(summary_grade['Grade_Value']):
        grade = 'Excellent' if v >= 3.5 else 'Good' if v >= 2.5 else 'Fair' if v >= 1.5 else 'Poor'
        ax4.text(i, 0.5, f"({grade})", ha='center')

    plt.tight_layout()
    plt.savefig('safety_benchmark_comparison.png', dpi=300, bbox_inches='tight')

    return "Visualizations created: safety_benchmark_comparison.png and safety_heatmap.png"

# Execute the benchmark visualization
df = parse_benchmark_data()
result = create_benchmark_visualizations(df)
print(result)
````

## **Original** Analytical Findings

Through comparative analysis of 12 safety metrics (VCR=Violent Content, CSE= Child Sexual Exploitation, etc.) across three models, we identified:

**Safety Capacity Trade-off Phenomenon**  
When models demonstrate strong performance in "defensive safety" dimensions like Privacy Protection (PRV) and Intellectual Property (IPV), their capabilities in "proactive safety" domains such as Hazardous Materials Handling (HTE) and Sexual Content Moderation (SXC) show significant degradation (Pearson r = -0.98). This pattern is particularly evident in Mistral-7B:

- PRV/IPV Scores: Good (9/10)  
- HTE/SXC Scores: Poor (5/10)

This reveals a **Safety Capability Ceiling Effect** in current LLMs - excessive optimization of specific safety dimensions appears to consume disproportionate reasoning resources for other safety mechanisms.



````python
# Correlation Calculation
from scipy.stats import pearsonr

defensive_scores = [9, 8, 7]  # PRV/IPV metrics
proactive_scores = [5, 6, 8]  # HTE/SXC metrics
r_coeff, p_value = pearsonr(defensive_scores, proactive_scores)
print(f"Pearson r: {r_coeff:.2f} (p={p_value:.4f})")
````
